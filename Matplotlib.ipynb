# 1 Installation et importation du package

```python
# Importation des packages
import matplotlib.pyplot as plt
import numpy as np
```

# 2 Tracé de graphiques simples avec Matplotlib

```python
# Génération de données
x=np.linspace(0, 10, 1000)
x
y=np.exp(x)
y
```

```python
# Tracer un graphique simple
plt.plot(x,y)
```

# 3 Personnalisation de graphiques avec Matplotlib

```python
# Changer la taille d'un graphique avec la largeur et la hauteur de la fihure en pouce
#plt.figure(figsize=(5,5)) #pour changer la taille du graphique
plt.plot(x,y)
```

```python
# Titre et axe
plt.plot(x,y)
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")           
```

```python
# Changer de couleur 
plt.plot(x,y,c="red")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")    
```

```python
# Changer le type de ligne
plt.plot(x,y,c="red",ls="--")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")    
```

```python
# Ajouter une légende
plt.plot(x,y,c="red",ls="-",label="exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")    
plt.legend()
```

```python
# Ajouter une grille sur le graphique
plt.plot(x,y,c="red",ls="-",label="exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")    
plt.legend()
plt.grid()
```

# 4 Faire plusieurs graphiques

## 4.1 Faire des graphiques sur un même carré

```python
# créer un graphique vide
z=2*y
y3=z*2
plt.plot(x,y,c="red",ls="-",label="exp")
plt.plot(x,z,c="green",ls="-",label="2*exp")
plt.plot(x,y3,c="blue",ls="-",label="4*exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")    
plt.legend()
```

```python
#Changer les limites des axes
z=2*y
y3=z*2
plt.plot(x,y,c="red",ls="-",label="exp")
plt.plot(x,z,c="green",ls="-",label="2*exp")
plt.plot(x,y3,c="blue",ls="-",label="4*exp")
plt.xlim((6,10)) # pour limiter les axes
plt.ylim((1000,80000))
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")    
plt.legend()
```

## 4.2 Faire des subplots

### 4.2.1 Approche fonctionnelle

```python
#Graphique avec subplot
plt.subplot(2,2,1)
plt.plot(x,y,c="red",ls="-",label="exp")
plt.subplot(2,2,2)
plt.plot(x,y,c="blue",ls="-",label="exp")
plt.subplot(2,2,3)
plt.plot(x,y,c="green",ls="-",label="exp")
plt.subplot(2,2,4)
plt.plot(x,y,c="black",ls="-",label="exp")
plt.subplots_adjust(wspace=0.5)  # pour rajouter de l'espace entre les graphiques
```

## 4.2.2 Approche orientée objet

```python
#Subplots
fig,axes=plt.subplots(2,1)
axes[0].plot(x,y,c="red")
axes[1].plot(x,y,c="blue")
```
